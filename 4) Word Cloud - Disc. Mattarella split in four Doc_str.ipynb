```python
# Add the word to the set of stop words. Use lowercase!
nlp.Defaults.stop_words.add('e')
nlp.vocab['e'].is_stop = True
nlp.Defaults.stop_words.add('.')
nlp.vocab['.'].is_stop = True
nlp.Defaults.stop_words.add(',')
nlp.vocab[','].is_stop = True
nlp.Defaults.stop_words.add(' ')
nlp.vocab[' '].is_stop = True
nlp.Defaults.stop_words.add('-')
nlp.vocab['-'].is_stop = True
nlp.Defaults.stop_words.add(':')
nlp.vocab[':'].is_stop = True
nlp.Defaults.stop_words.add(';')
nlp.vocab[';'].is_stop = True
```

```python
# Transforming text into a token sequence
token_list1 = []
token_list2 = []
token_list3 = []
token_list4 = []
for token in doc1:
    token_list1.append(token.text)
for token in doc2:
    token_list2.append(token.text)
for token in doc3:
    token_list3.append(token.text)
for token in doc4:
    token_list4.append(token.text)
```
